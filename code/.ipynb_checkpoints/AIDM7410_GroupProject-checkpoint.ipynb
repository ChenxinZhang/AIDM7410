```python
def getandsaveText(filepaths):
    """
    filepaths::Picture address list
    
    """
    # Get the text of each picture
    request_url = "https://aip.baidubce.com/rest/2.0/ocr/v1/webimage"
        # Open picture file in binary mode
    for filepath in filepaths:
        if '.jp' in filepath:           
            try:
                f = open(filepath, 'rb')
                print("open", filepath)
                img = base64.b64encode(f.read())
                params = {"image":img}
                access_token = '24.a155b276480ff9462d0597f43658056a.2592000.1622115233.282335-24078709'
                request_url = request_url + "?access_token=" + access_token
                headers = {'content-type': 'application/x-www-form-urlencoded'}
                # The image has recognized the returned text
                response = requests.post(request_url, data=params, headers=headers)
                results = response.json()
                #wordsNum = results['words_result_num']
                words = [i['words'] for i in results['words_result']]
                #words.append(wordsNum)
                # Save to a CSV file
                mistake_count = 0
                filepath = filepath.split('.',-1)
                if len(filepath) == 3:
                    filepath = filepath[0]+filepath[1]
                elif len(filepath) == 2:
                    filepath = filepath[0]
                else:
                    mistake_count = mistake_count + 1
                resultsText = pd.DataFrame(words)
                resultsText.to_csv(filepath +'.csv',encoding = "utf-8")
                print(filepath," Saved after text recognition")
                time.sleep(1)
            except BaseException:
                raise("something wrong")
                pass
          
        else:
            pass
        
    print("There may be omissions",mistake_count)
getandsaveText(cosmoWomen)
getandsaveText(chineseWomen)
getandsaveText(forhimMagazineWomen)
```

```python
# Get the path of the CSV file
def listdir(path, list_name):  # Incoming stored list
    """
    path: root path
    list_name::List to store text paths
    """
    for file in os.listdir(path):
        print("file",file)
        file_path = os.path.join(path,file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)
```

```python
# Read all the contents of a magazine
def getContent(filepaths):
    dfg = pd.DataFrame()
    for path in filepaths:
        if ".csv" in path:
            try:
                #print(path)
                dfg_wait = pd.read_csv(path)
                #print(dfg_wait)
                dfg = dfg.append(dfg_wait)
            except:
                pass
        else:
            pass
    #print(dfg)
    return dfg                
```

```python
def modify_filenames(upperpath,filepathlist):
    """
    upperpath::需要修改的文件或者文件夹名的上一层路径
    filepathlist::需要修改的文件名列表
    
    """
    for path in filepathlist:
        oldname=upperpath+ os.sep + path
        print(oldname)
        #设置新文件名
        filename = ''
        #提取原文件名中的数字
        filenames = re.findall(r"\d+\.?\d*",path)
        for index in range(len(filenames)):
            filename = filename+filenames[index]    
        newname=upperpath + os.sep +filename
        os.rename(oldname,newname)   #用os模块中的rename方法对文件改名
        print(oldname,'======>',newname)
        print(filename)
modify_filenames(cosmopath,cosmopathlist)
modify_filenames(forhimmagazinepath,forhimmagazinepathlist)
modify_filenames(chinesewomenpath,chinesewomenpathlist)
```

```python
# 定义图像拼接函数
def image_compose():
    IMAGE_COLUMN = 10
    IMAGE_ROW = len(image_names_chinesewomen)//IMAGE_COLUMN+1
    to_image = Image.new('RGB', (IMAGE_COLUMN  * IMAGE_SIZE, IMAGE_ROW * IMAGE_SIZE)) 
    # 创建一个新图
    # 循环遍历，把每张图片按顺序粘贴到对应位置上
    x = 1 # rows
    y = 0 # columns
    for index in range(len(image_names_chinesewomen)):
        try:
            from_image = Image.open(image_names_chinesewomen[index]).resize(
                (IMAGE_SIZE+1, IMAGE_SIZE+1), Image.ANTIALIAS)
            indexcolumns = index//10
            index = index - indexcolumns*10
            to_image.paste(from_image, (index * IMAGE_SIZE, indexcolumns * IMAGE_SIZE))
            y = y + 1
            
        except:
            pass
    to_image.save(image_names_chinesewomen[0],'png')
    print(image_names_chinesewomen[0],'saved')
    #save(IMAGE_SAVE_PATH,'jpeg')  
    # 保存新图
image_compose()  # 调用函数
```

```python
# 获取每个图片的RGB平均值
def findAvgRgb(imagePath):
    im = Image.open(imagePath)
    im = im.convert('RGB')
    r, g, b = im.getpixel((1,1))
    rgb = (r, g, b)
    avg = 0
    avg = (rgb[0]+rgb[1]+rgb[2])/3
    return avg
def take1(elem):
    return elem[0]
```
