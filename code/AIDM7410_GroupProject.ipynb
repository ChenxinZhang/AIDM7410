```python
def getandsaveText(filepaths):
    """
    filepaths::Picture address list
    
    """
    # Get the text of each picture
    request_url = "https://aip.baidubce.com/rest/2.0/ocr/v1/webimage"
        # Open picture file in binary mode
    for filepath in filepaths:
        if '.jp' in filepath:           
            try:
                f = open(filepath, 'rb')
                print("open", filepath)
                img = base64.b64encode(f.read())
                params = {"image":img}
                # Enter the access_token of registered Baidu AI platform applications
                access_token = access_token
                request_url = request_url + "?access_token=" + access_token
                headers = {'content-type': 'application/x-www-form-urlencoded'}
                # The image has recognized the returned text
                response = requests.post(request_url, data=params, headers=headers)
                # Obtain the JSON formatted data composed of the recognized text
                results = response.json()
                words = [i['words'] for i in results['words_result']]
                # Save to a CSV file
                mistake_count = 0
                filepath = filepath.split('.',-1)
                if len(filepath) == 3:
                    filepath = filepath[0]+filepath[1]
                elif len(filepath) == 2:
                    filepath = filepath[0]
                else:
                    mistake_count = mistake_count + 1
                resultsText = pd.DataFrame(words)
                # Save to the previous folder respectively
                resultsText.to_csv(filepath +'.csv',encoding = "utf-8")
                print(filepath," Saved after text recognition")
                time.sleep(1)
            except BaseException:
                raise("something wrong")
                pass
          
        else:
            pass
        
    print("There may be omissions",mistake_count)
```

```python
def modify_foldernames(upperpath,filepathlist):
    """
    upperpath::The upper path of the file or folder name to be modified
    filepathlist::List of file names to modify
    
    """
    for path in filepathlist:
        oldname=upperpath+ os.sep + path
        # print(oldname)
        #Set new file names
        filename = ''
        # Extract the number in the original file name
        filenames = re.findall(r"\d+\.?\d*",path)
        for index in range(len(filenames)):
            filename = filename+filenames[index]    
        newname=upperpath + os.sep +filename
        if newname == oldname:
            pass
            # print("The name has been changed")
        else:
            os.rename(oldname,newname)   
            print(" Renaming ")
            print(oldname,'======>',newname)
        
```

```python
# Define image mosaic function
def image_compose(image_names):
    
    IMAGE_COLUMN = 10
    IMAGE_ROW = len(image_names)//IMAGE_COLUMN+1
    to_image = Image.new('RGB', (IMAGE_COLUMN  * IMAGE_SIZE, IMAGE_ROW * IMAGE_SIZE)) 
    # Create a new diagram
    # Loop through and paste each picture into the corresponding position in order
    x = 1 # rows
    y = 0 # columns
    for index in range(len(image_names)):
        try:
            from_image = Image.open(image_names[index]).resize(
                (IMAGE_SIZE+1, IMAGE_SIZE+1), Image.ANTIALIAS)
            indexcolumns = (index-1)//10
            index = index - indexcolumns*10
            to_image.paste(from_image, ((index-1) * IMAGE_SIZE, indexcolumns * IMAGE_SIZE))

        except:
            pass
    if index < IMAGE_COLUMN:
        supplement_image = Image.new('RGB', ((IMAGE_COLUMN-index)*IMAGE_SIZE, IMAGE_SIZE),'white')
        to_image.paste(supplement_image, (index * IMAGE_SIZE, indexcolumns * IMAGE_SIZE))
    to_image.save(image_names[0],'png')
    print(image_names[0],'saved')
```

```python
# Organize the main color maps of the magazine by year
def createColorPicture(upperPathList,imagePathList,numColor):
    """
    upperPathList::Saved path
    imagePathList::Set of image absolute paths to be taken
    numColor:: Int type, how many colors to extract?
    
    """
    imagePathList1 = []
    for image in imagePathList:
        if '.jp' in image:
            imagePathList1.append(image)
    imagePathList = imagePathList1
    
    for index in range(len(imagePathList)):        
        img_path = imagePathList[index]
        if '.jp' in img_path:
            #print("open:",img_path)
            image = Image.open(img_path)
            # Number of main colors to extract
            small_image = image.resize((80, 80))
            result = small_image.convert('P', palette=Image.ADAPTIVE, colors=numColor)   # image with 5 dominating colors
            result = result.convert('RGB')            
            main_colors = result.getcolors(80*80)
            counts = 0
            for i in range(len(main_colors)):
                 counts = counts + main_colors[i][0]
            for count, col in main_colors:
                if count < 40:
                    continue
                a = np.zeros((224,224,3))
                a = a + np.array(col)
                filepath = upperPathList + str(count)+'_'+ str(index) + '.png'
                # print("now upperPathList is ",upperPathList)
                cv2.imwrite(filepath,a.astype(np.uint8)[:,:,::-1])
                #print(filepath,'is saved first')
                img = cv2.imread(filepath)
                height, width = img.shape[:2]  
                # Set image size by percentage
                size = (int(width*(count/counts)), int(height))  
                shrink = cv2.resize(img, size, interpolation=cv2.INTER_AREA)
                cv2.imwrite(filepath,shrink) 
                #print(filepath,"is saved second")
                cv2.waitKey()
```

```python
# Get the average RGB value of each picture
def findAvgRgb(imagePath):
    im = Image.open(imagePath)
    im = im.convert('RGB')
    r, g, b = im.getpixel((0,1))
    rgb = (r, g, b)
    avg = 0
    avg = (rgb[0]+rgb[1]+rgb[2])/3
    return avg
def take1(elem):
    return elem[0]
# Define image mosaic function
def image_compose_color(image_names,year):   
    """
    year::year
    image_names::lists of picture files by year
    """
    image_names_filter = []
    for name in image_names:
        if 'png' in name:
            image_names_filter.append(name)
        else:
            pass
        #sort
    image_names_filter_sorted = []
    for image in image_names_filter:
        avg = findAvgRgb(image)
        image_names_filter_sorted.append([avg,image])

    image_names_filter_sorted.sort()#(key = take1)
    to_image = Image.new('RGB', (224*12, 224)) 
    width = 0
    for index in range(len(image_names_filter_sorted)):
        #print("Now open this picture:",image_names_filter_sorted[index][1])
        try:
            from_image = Image.open(image_names_filter_sorted[index][1])
            addwidth = from_image.size[0]
            to_image.paste(from_image, (width, 0))
            width = width + addwidth
        except:
            pass
    if width<224*12:
        supplement_image = Image.new('RGB', (224*12-width, 224),'white')
        to_image.paste(supplement_image, (width, 0))
    savedpath = 'C:/Users/Cheung/Downloads/'+ image_names_filter_sorted[index][1].split('/')[-3]+'/'
    yearImagePath = savedpath+ str(year)+ '.png'
    to_image.save(yearImagePath,'png')

    print(yearImagePath,'saved')
```
